### EDA

Check distribution of LIWC results

```python
quantiles_all = liwc_results_filtered2[["emo_pos", "emo_neg", "emo_anx", 
                              "emo_sad", "emo_anger", "moral"]] \
                            .quantile([0.1, 0.25, 0.5, 0.75, 0.9])
quantiles_all
```

Check correlation between LIWC results

```python
# Compute the correlation matrix
corr_matrix = liwc_results_filtered2[["emo_pos", "emo_neg", "emo_anx", 
                              "emo_sad", "emo_anger", "moral"]].corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

# Show the plot
plt.title("Heatmap of LIWC Results")
plt.show()
```
